```python
# BATTING POINTS
# run = 1 * / boundary B = 1 * / six B = 2 *30 /run B = 4 @ /Half-century B = 8 @ /Century B = 16 @ /Duck = -2 @

# SR:
# >170 = 6 @# 150-170 = 4 @# 130-150 = 2 @# 60-70 = -2 @# 50-59.00 = -4 @# <50 = -6 @

# BOWLING POINTS# wicket = 25 * # B(LBW/Bowled) = 8 *# 3 w = 4 @# 4 w = 8 @# 5 w = 16 @# maiden = 12 @

# eco:# <5 = 6 @# 5-5.99 = 4 @# 6-7 = 2 @# 10-11 = -2 @# 11.01-12 = -4 @# >12 = -6 @

# FIELDING POINTS
# catch = 8# 3 catch B = 4# stumping = 12# RO(direct hit) = 12 # RO = 6
```
